```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
 #подключение библиотек
```

```python
df = pd.read_csv("vgsales_15.csv") #загрузка датасета
df
```

```python
from sklearn.preprocessing import OrdinalEncoder

Oenc = OrdinalEncoder() # преобразовываем текстовые значения в числовой формат
df [['Name', 'Platform', 'Year', 'Publisher']] = Oenc.fit_transform(df[['Name', 'Platform', 'Year', 'Publisher']])
df

```

```python
#df.to_csv('vgsales_15_new.csv', index=False)
```

```python
from sklearn.preprocessing import OneHotEncoder

OHenc = OneHotEncoder() #горячей кодировкой бинарно указываем на принадлежность к категории
df['Genre'] = OHenc.fit_transform(df[['Genre']]).toarray()
df
```

```python
from sklearn.model_selection import train_test_split

Y = df['Genre'] #обучающие данные
X = df.drop(columns='Genre') #тестовые данные

x_train, x_test, y_train, y_test = train_test_split(X,Y) #выборка с обучающими и тестовыми признаками/цел. переменной
```

```python
x_train = x_train.dropna()
y_train = y_train[x_train.index] #удаление пустых значений и перезапись
```

```python
x_test = x_test.dropna()
y_test = y_test[x_test.index]
```

```python
from sklearn.tree import DecisionTreeClassifier #подключение класса 
dtc = DecisionTreeClassifier()
dtc = dtc.fit(x_train, y_train) #обучение первой модели
dtc_plot = DecisionTreeClassifier() #модель для визуализации
dtc_plot = dtc_plot.fit(x_train.iloc[::10], y_train.iloc[::10]) #обучение каждой 10 выборки
```

```python
from sklearn import tree
plt.figure(figsize=(36,16)) #построение дерева
tree.plot_tree(dtc_plot, filled=True, fontsize=10)
```

```python
from sklearn import metrics
Y_predict = dtc.predict(x_test) #прогнозы тестовых дынных
con_mat = metrics.confusion_matrix(y_test, Y_predict) #матрица ошибок
sns.heatmap(con_mat, annot=True, fmt="d") #отрисовка тепловой картой
plt.xlabel("Предсказано") #подписи к графику
plt.ylabel("Реально")
```

```python
print("Точность", metrics.accuracy_score(y_test, Y_predict)) #правильность 
print("Чувствительность", metrics.recall_score(y_test, Y_predict)) #полнота
print("Аккуратность", metrics.precision_score(y_test, Y_predict)) #правильность предсказания + среди всех +
print("Финальная", metrics.f1_score(y_test, Y_predict)) #общая оценка
```

1. ~70% - правильно определённых наблюдений от общего числа наблюдений, отражающее высокую производительность модели
2. ~43% - полноты модели, что говорит о среднем обнаружении положительных случаев
3. ~35% - доля правильно определённых положительных случаев относительно всех положительных для модели случаев
4. ~39% - зреднее значение баланса точночтти и чувствительности

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

Y = df['Genre']
X = df.drop(columns='Genre')

x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.2, random_state=42)

x_train = x_train.dropna()
y_train = y_train[x_train.index]
x_test = x_test.dropna()
y_test = y_test[x_test.index]

dtc = DecisionTreeClassifier()
dtc = dtc.fit(x_train, y_train)
dtc_plot = DecisionTreeClassifier(max_depth=6, min_samples_leaf=4, max_leaf_nodes=200) #дерево с гиперпараметрами (глубина уровней,образцы,общее колво узлов)
dtc_plot = dtc_plot.fit(x_train.iloc[::10], y_train.iloc[::10])

plt.figure(figsize=(36,16))
tree.plot_tree(dtc_plot, filled=True, fontsize=10)
```

```python
from sklearn import metrics
Y_predict = dtc.predict(x_test)
con_mat = metrics.confusion_matrix(y_test, Y_predict)
sns.heatmap(con_mat, annot=True, fmt="d")
plt.xlabel("Предсказано")
plt.ylabel("Реально")
```

```python
print("Точность", metrics.accuracy_score(y_test, Y_predict))
print("Чувствительность", metrics.recall_score(y_test, Y_predict))
print("Аккуратность", metrics.precision_score(y_test, Y_predict))
print("Финальная", metrics.f1_score(y_test, Y_predict))
```

1. ~66% - правильно определённых наблюдений от общего числа наблюдений, отражающее высокую производительность модели
2. ~31% - полноты модели, что говорит о среднем обнаружении положительных случаев
3. ~20% - доля правильно определённых положительных случаев относительно всех положительных для модели случаев
4. ~24% - зреднее значение баланса точночтти и чувствительности
